```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from urllib.request import urlopen
import pandas as pd
import time
import json
```

```python
driver = webdriver.Chrome()  
```

```python
driver.get('https://undi.info/')
```

Scraping Parliament Election History Data

```python
state_btns = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div')
for state_btn in state_btns:
    state_btn.click()
    time.sleep(1)

state_codes = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div/label/table/tbody/tr/td[1]/img')
state_codes = [x.get_attribute('src') for x in state_codes]
state_codes = [x.replace('https://undi.info/state_flags/','') for x in state_codes]
state_codes = [x.replace('.png','') for x in state_codes]

seat_codes_df = pd.DataFrame()
for i in range(len(state_codes)):
    pseat_btns = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div['+str(i+1)+']/div/table/tbody/tr/td[2]/div')
    seat_names = [x.text for x in pseat_btns]
    df = pd.DataFrame()
    df['seat_name'] = seat_names
    df['state_code'] = state_codes[i]
    seat_codes_df = pd.concat([seat_codes_df,df],axis=0)
seat_codes_df.reset_index(drop=True,inplace=True)

pReqURLs = 'https://api.undi.info/election?seat='+seat_codes_df.state_code+'.p.'+seat_codes_df.seat_name
pReqURLs = [x.replace(' ','%20') for x in pReqURLs]
pReqURLs
```

```python
driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[1]/div[2]/div/div[1]/div[1]').click()

seat_codes_df = pd.DataFrame()
for i in range(len(state_codes)):
    pseat_btns = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div['+str(i+1)+']/div/table/tbody/tr/td[2]/div')
    seat_names = [x.text for x in pseat_btns]
    df = pd.DataFrame()
    df['seat_name'] = seat_names
    df['state_code'] = state_codes[i]
    seat_codes_df = pd.concat([seat_codes_df,df],axis=0)
seat_codes_df.reset_index(drop=True,inplace=True)

ReqURLs = 'https://api.undi.info/election?seat='+seat_codes_df.state_code+'.p.'+seat_codes_df.seat_name
ReqURLs = [x.replace(' ','%20') for x in ReqURLs]
pReqURLs = list(set(pReqURLs+ReqURLs))
pReqURLs
```

Scraping Dun Election History Data

```python
pd_button = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[1]/div[2]/div/div[2]/div')
pd_button[1].click()

state_codes = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div/label/table/tbody/tr/td[1]/img')
state_codes = [x.get_attribute('src') for x in state_codes]
state_codes = [x.replace('https://undi.info/state_flags/','') for x in state_codes]
state_codes = [x.replace('.png','') for x in state_codes]

seat_codes_df = pd.DataFrame()
for i in range(len(state_codes)):
    pseat_btns = driver.find_elements(By.XPATH,'/html/body/div[1]/div/div/div[2]/div[2]/div[2]/div['+str(i+1)+']/div/table/tbody/tr/td[2]/div')
    seat_names = [x.text for x in pseat_btns]
    df = pd.DataFrame()
    df['seat_name'] = seat_names
    df['state_code'] = state_codes[i]
    seat_codes_df = pd.concat([seat_codes_df,df],axis=0)
seat_codes_df.reset_index(drop=True,inplace=True)

dReqURLs = 'https://api.undi.info/election?seat='+seat_codes_df.state_code+'.n.'+seat_codes_df.seat_name
dReqURLs = [x.replace(' ','%20') for x in dReqURLs]
dReqURLs
```

```python
driver.close()
```

Get Parlimen Data

```python
parlimen_df = pd.DataFrame()
for reqURL in pReqURLs:
    response = None
    while response is None:
        try:
            response = urlopen(reqURL)
        except Exception as e:
            print('URL Error:'+reqURL+'\n'+str(e))
            time.sleep(1)
            response = None
    data_json = json.loads(response.read())
    
    for year_key in data_json.keys():
        df = pd.DataFrame(data_json[year_key]['candidates'])
        df['Event'] = year_key
        
        for key in data_json[year_key].keys():
            if type(data_json[year_key][key]) is str:
                df[key] = data_json[year_key][key]
            if key=='info':
                for info_key in data_json[year_key][key].keys():
                    df[info_key] = data_json[year_key][key][info_key]
        
        parlimen_df = pd.concat([parlimen_df,df],axis=0)
parlimen_df.reset_index(drop=True,inplace=True)
```
