```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from lxml import etree
import re
```

```python
# read html data
URL = "https://en.wikipedia.org/wiki/Results_of_the_2013_Malaysian_general_election_by_parliamentary_constituency"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
dom = etree.HTML(str(soup))
```

```python
# scrape states
states = []
for i in range(2,18):
    states.append(dom.xpath('/html/body/div[3]/div[3]/div[5]/div[1]/h2['+str(i)+']/span[1]')[0].text)
states
```

```python
# scrape tables
tables = soup.find_all('table', class_='wikitable sortable')

election_df = pd.DataFrame()
for i in range(1,len(tables)):
    table = tables[i]
    
    columns_name = ['#','Constituency',
                    'Winner','Winner_Votes','Winner_Votes%',
                    'Opponents','Opponents_Votes','Opponents_Votes%',
                    'Majority','Incumbent','Eligible_Voters',
                    'Malay_Voters%','Chinese_Voters%','Indian_Voters%','Others_Voters%',
                    'Muslim_Bumi_Voters%','Non_Muslim_Bumi_Voters%',
                    'Voter_Turnout','Voter_Turnout%','Spoilt_Votes','Spoilt_Votes%'
                   ]
    
    df = pd.DataFrame()
    for row in table.tbody.find_all('tr'):
        columns = row.find_all('td')
        
        json = {}
        if len(columns)==len(columns_name):
            for j in range(len(columns)):
                json[columns_name[j]] = columns[j].text.strip()
        
            df = df.append(json, ignore_index=True)
        elif len(columns)==3:
            json = df.tail(1)
            
            for j in range(len(columns)):
                json[columns_name[j+5]] = columns[j].text.strip()
            
            df = df.append(json, ignore_index=True)
    df['State'] = states[i-1]
    election_df = election_df.append(df, ignore_index=True)
```

```python
election_df
```

```python
winner_df = election_df.iloc[:, [0,1,2,3,4]]
winner_df = winner_df.drop_duplicates()
winner_df.reset_index(drop=True,inplace=True)
winner_df.columns = ['#','Constituency','Candidate','Votes','Votes%']
winner_df.head()
```

```python
loser_df = election_df.iloc[:,[0,1,5,6,7]]
loser_df = loser_df.drop_duplicates()
loser_df.reset_index(drop=True,inplace=True)
loser_df.columns = ['#','Constituency','Candidate','Votes','Votes%']
loser_df.head()
```

```python
candidates_df = pd.concat([winner_df,loser_df], axis=0)
candidates_df.reset_index(drop=True,inplace=True)
candidates_df
```

```python
temp = [re.search('\\([a-zA-Z]+[]?[a-zA-Z]*\\)',x).group() 
 if re.search('\\([a-zA-Z]+[]?[a-zA-Z]*\\)',x) is not None 
 else '' for x in candidates_df.Candidate]
temp = [x.split('') for x in temp]

coalition = [x[0] if len(x)==2 else '' for x in temp]
party = [x[len(x)-1] for x in temp]

coalition = [''.join(e for e in x if e.isalnum()) for x in coalition]
party = [''.join(e for e in x if e.isalnum()) for x in party]

candidates_df['Coalition'] = coalition
candidates_df['Party'] = party
candidates_df['Candidate'] = [re.sub('\\([a-zA-Z]+[]?[a-zA-Z]*\\)','',x) for x in candidates_df.Candidate]

candidates_df = candidates_df.sort_values('#').reset_index(drop=True)
```

```python
candidates_df
```

```python
constituency_df = election_df.drop(['Winner','Winner_Votes','Winner_Votes%',
                                    'Opponents','Opponents_Votes','Opponents_Votes%'],axis=1)
constituency_df = constituency_df.drop_duplicates()
constituency_df
```

```python
temp = [re.search('\\([a-zA-Z]+[]?[a-zA-Z]*\\)',x).group() 
 if re.search('\\([a-zA-Z]+[]?[a-zA-Z]*\\)',x) is not None 
 else '' for x in constituency_df.Incumbent]
temp = [x.split('') for x in temp]

coalition = [x[0] if len(x)==2 else '' for x in temp]
party = [x[len(x)-1] for x in temp]

coalition = [''.join(e for e in x if e.isalnum()) for x in coalition]
party = [''.join(e for e in x if e.isalnum()) for x in party]

constituency_df['Incumbent_Coalition'] = coalition
constituency_df['Incumbent_Party'] = party
constituency_df['Incumbent'] = [re.sub('\\([a-zA-Z]+[]?[a-zA-Z]*\\)','',x) for x in constituency_df.Incumbent]

constituency_df
```

```python
all_df = pd.merge(candidates_df,constituency_df,how='left')
all_df = all_df.iloc[:,[x for x in range(0,2)] + [20] + [x for x in range(2,9)] + [21,22] + [x for x in range(9,20)]]
all_df
```

```python
all_df.columns.to_list()
```

```python
all_df.replace('%','', regex=True,inplace=True)
all_df.replace(',','', regex=True,inplace=True)
```

```python
for column in all_df.columns:
    try:
        all_df[[column]] = all_df[[column]].apply(pd.to_numeric)
    except Exception as e:
        print(e)
```
